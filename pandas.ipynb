```python
import datetime
import random

from pyecharts import options as opts
from pyecharts.charts import Calendar


def calendar_base() -> Calendar:
    begin = datetime.date(2017, 1, 1)
    end = datetime.date(2017, 12, 31)
    data = [
        [str(begin + datetime.timedelta(days=i)), random.randint(1000, 25000)]
        for i in range((end - begin).days + 1)
    ]

    c = (
        Calendar()
        .add("", data, calendar_opts=opts.CalendarOpts(range_="2017"))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Calendar-2017年微信步数情况"),
            visualmap_opts=opts.VisualMapOpts(
                max_=20000,
                min_=500,
                orient="horizontal",
                is_piecewise=True,
                pos_top="230px",
                pos_left="100px",
            ),
        )
    )
    return c
```

```python
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Page, Pie


def pie_base() -> Pie:
    c = (
        Pie()
        .add("", [list(z) for z in zip(Faker.choose(), Faker.values())])
        .set_global_opts(title_opts=opts.TitleOpts(title="Pie-基本示例"))
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
    )
    return c

i=pie_base()
i.render_notebook()
```

```python
import json
import os

from pyecharts import options as opts
from pyecharts.charts import Page, Sankey


def sankey_base() -> Sankey:
    nodes = [
        {"name": "成本"},
        {"name": "增值税进项"},
        {"name": "收入（不含增值税）"},
        {"name": "增值税销项"},
        {"name": "增值税"},
        {"name": "毛利润"},
        {"name": "管理费用"},
        {"name": "净利润"},
        {"name": "所得税"},
    ]

    links = [
        {"source": "收入（不含增值税）", "target": "成本", "value": 100},
        {"source": "收入（不含增值税）", "target": "毛利润", "value": 20},
        {"source": "毛利润", "target": "净利润", "value": 20},
        {"source": "增值税销项", "target": "增值税进项", "value": 17},
        {"source": "增值税进项", "target": "增值税", "value": 10},
        {"source": "毛利润", "target": "管理费用", "value": 7},
        {"source": "毛利润", "target": "所得税", "value": 3},
    ]
    c = (
        Sankey()
        .add(
            "sankey",
            nodes,
            links,
            linestyle_opt=opts.LineStyleOpts(opacity=0.2, curve=0.5, color="source"),
            label_opts=opts.LabelOpts(position="right"),
        )
        .set_global_opts(title_opts=opts.TitleOpts(title="税费构成"))
    )

    return c
san = sankey_base()
san.render_notebook()
```
